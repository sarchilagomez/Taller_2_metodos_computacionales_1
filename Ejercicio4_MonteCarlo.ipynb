1. Gráfica de error porcentual

```python
def Funcion(x):
        return np.exp(-x)*np.sin(x)

Iteo = 0.5*(1+np.exp(-np.pi))

a = 0
b = np.pi

Errores = []
Ns = []
for i in range(10, 10**5, 1000):
    N = i
    x = np.random.uniform(a,b,N) #N es el número de muestras aleatorias que tomará

    fi = Funcion(x)

    I = (b-a)*sum(fi)/N
    error = np.abs(1-I/Iteo)
    Errores.append(error)
    Ns.append(N)

plt.plot(Ns, Errores)
plt.xlabel("N")
plt.ylabel("Error")
plt.title("Error vs N en Montecarlo")
plt.show()
    
```

2. Aproximar valor de la integral

```python
R = 1
N = 10000
x = np.random.uniform(-R,R,N)
y = np.random.uniform(-R,R,N)
z = np.random.uniform(-R,R,N)

suma = 0

for i in range(N):
    if x[i]**2+y[i]**2+z[i]**2 <= 1:
        suma += np.sin(x[i]**2 + y[i]**2 + z[i]**2)*np.e**(x[i]**2 + y[i]**2 + z[i]**2)  #Se suma uno puesto que la función es 1

I = (8*suma)/N
#Se coloca 8 dado que el volumen del cubo es 8 
```

```python
print("El valor aproximado de la integral es:", I)
```
